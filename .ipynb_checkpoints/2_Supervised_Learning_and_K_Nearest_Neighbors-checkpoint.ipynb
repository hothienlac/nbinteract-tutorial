```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
from collections import defaultdict
from sklearn.preprocessing import LabelEncoder
d = defaultdict(LabelEncoder)
```

```python
data = pd.read_csv('data/Orange_Telecom_Churn_Data.csv')
data.head()
```

## Question 2

* Notice that some of the columns are categorical data and some are floats. These features will need to be numerically encoded using one of the methods from the lecture.
* Finally, remember from the lecture that K-nearest neighbors requires scaled data. Scale the data using one of the scaling methods discussed in the lecture.

```python
fit = data.apply(lambda x: d[x.name].fit_transform(x))
fit.head()
```

```python
# Decode

fit.apply(lambda x: d[x.name].inverse_transform(x)).head()
```

```python
scaled = fit.apply(lambda x: x/max(x))
scaled.head()
```

## Question 3

* Separate the feature columns (everything except `churned`) from the label (`churned`). This will create two tables.
* Fit a K-nearest neighbors model with a value of `k=3` to this data and predict the outcome on the ``same data``.

```python
churned = dict(tuple(scaled.groupby('churned')))
not_churned = churned[0]
churned = churned[1]
churned.head()
not_churned.head()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(scaled, test_size=0.2)
x_train = train.drop(['churned'], axis = 1)
y_train = train.churned
x_test = test.drop(['churned'], axis = 1)
y_test = test.churned

from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors = 3)
model.fit(x_train, y_train)
predicted = model.predict(x_test)
predicted[:10]
```

## Question 4

Ways to measure error haven't been discussed in class yet, but accuracy is an easy one to understand--it is simply the percent of labels that were correctly predicted (either true or false). 

* Write a function to calculate accuracy using the actual and predicted labels.
* Using the function, calculate the accuracy of this K-nearest neighbors model on the data.

```python
a = tuple(zip(y_test, predicted))
check = [1 if (i[0] == i[1]) else 0 for i in a]
accuracy = np.mean(check)
accuracy
```

## Question 5

* Fit the K-nearest neighbors model again with `n_neighbors=3` but this time use distance for the weights. Calculate the accuracy using the function you created above. 
* Fit another K-nearest neighbors model. This time use uniform weights but set the power parameter for the Minkowski distance metric to be 1 (`p=1`) i.e. Manhattan Distance.

When weighted distances are used for part 1 of this question, a value of 1.0 should be returned for the accuracy. Why do you think this is? *Hint:* we are predicting on the data and with KNN the model *is* the data. We will learn how to avoid this pitfall in the next lecture.

```python
weighted_model = KNeighborsClassifier(n_neighbors = 3, weights = 'distance')
weighted_model.fit(x_train, y_train)
predicted_2 = weighted_model.predict(x_test)
predicted_2[:10]
```

```python
a_2 = tuple(zip(y_test, predicted_2))
check_2 = [1 if (i[0] == i[1]) else 0 for i in a_2]
accuracy_2 = np.mean(check_2)
accuracy_2
```

## Question 6

* Fit a K-nearest neighbors model using values of `k` (`n_neighbors`) ranging from 1 to 20. Use uniform weights (the default). The coefficient for the Minkowski distance (`p`) can be set to either 1 or 2--just be consistent. Store the accuracy and the value of `k` used from each of these fits in a list or dictionary.
* Plot (or view the table of) the `accuracy` vs `k`. What do you notice happens when `k=1`? Why do you think this is? *Hint:* it's for the same reason discussed above.

```python
xmodel = list(range(200))
xaccuracy = list(range(200))
for i in range(200):
    xmodel[i] = KNeighborsClassifier(n_neighbors = i+1)
    xmodel[i].fit(x_train, y_train)
    xpredicted = xmodel[i].predict(x_test)
    xa = tuple(zip(y_test, xpredicted))
    xcheck = [1 if (i[0] == i[1]) else 0 for i in xa]
    xaccuracy[i] = np.mean(xcheck)

q = list(range(200))
q = [i+1 for i in q]
plt.plot(q, xaccuracy)
```
